```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("MLB_90s_to_10s.csv")
```

```python
df=df.drop(columns=["Unnamed: 0","Unnamed: 0.1","Lg","Pos Summary","Rk"])
df.head(3)
```

```python
def name_split(name):
    if "*" in name:
        name=name.split("*")
        return name[0]
    elif "#" in name:
        name=name.split("#")
        return name[0]
    else:
        name=name.split("\\")
        return name[0]
```

```python
df["Name"] = df.apply(lambda row: name_split(row["Name"]), axis=1)
df.head(5)
```

```python
df2=df
```

```python
len(df2)
```

```python
for i in range(0,len(df2)):
    try:
        if df2["Tm"][i]=="TOT":
            name=df2["Name"][i]

            if df2["Name"][i+4] == name:
                df2["Tm"][i]=df2["Tm"][i+4]
                df2=df2.drop(index=[i+1,i+2,i+3,i+4],axis=0)

            elif df2["Name"][i+3] == name:
                df2["Tm"][i]=df2["Tm"][i+3]
                df2=df2.drop(index=[i+1,i+2,i+3],axis=0)

            elif df2["Name"][i+2] == name:
                df2["Tm"][i]=df2["Tm"][i+2]
                df2=df2.drop(index=[i+1,i+2],axis=0)
    except KeyError:
        pass
            
df2.head(25)
```

```python
for i in range(0,len(df)):
    try:
        if np.isnan(df2["OBP"][i]):
            df2=df2.drop(index=i,axis=0)
    except KeyError:
        pass
df2.head(10)
```

```python
for i in range(0,len(df)):
    try:
        if df2["PA"][i]<150:
            df2=df2.drop(index=i,axis=0)
    except KeyError:
        pass
df2.head(10)
```
